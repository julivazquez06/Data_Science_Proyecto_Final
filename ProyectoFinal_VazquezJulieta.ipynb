**6. Feature Selection: reducir la dimensionalidad**

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import LabelEncoder
```

```python
# 1. Eliminar columnas no numéricas o que no aportan al modelo
Amazon_model = Amazon.drop(['Product_name', 'Category Name', 'ID_PRODUCT2', 'Hot_offer?'], axis=1)
```

```python
# 2. Convertir columnas categóricas a numéricas (si quedan)
Amazon_model = Amazon_model.apply(lambda col: LabelEncoder().fit_transform(col) if col.dtypes == 'object' else col)
```

```python
# 3. Separar variables predictoras (X) y variable objetivo (y)
X = Amazon_model.drop('Orders', axis=1)
y = Amazon_model['Sale_Price']
```

```python
# 4. Aplicar SelectKBest con f_regression
selector = SelectKBest(score_func=f_regression, k=8)  # elegir las 8 mejores características
```

```python
# 5. Ajustar selector y transformar X
X_selected = selector.fit_transform(X, y)
```

```python
# 6. Obtener las columnas seleccionadas
selected_features = X.columns[selector.get_support()]
```

```python
# 7. Crear un nuevo DataFrame con esas features
X_selected_df = pd.DataFrame(X_selected, columns=selected_features)
```

```python
# 8. Mostrar las features elegidas
print("Features seleccionadas:", selected_features.tolist())
```

```python
# 9. Mostrar el nuevo dataset reducido
print(X_selected_df.head())
```

7. Modelo de regresión o clasificación

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
# 1. Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y, test_size=0.2, random_state=42)
```

```python
# 2. Instanciar el modelo
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
# 3. Entrenar el modelo con los datos seleccionados
model.fit(X_train, y_train)
```

```python
# 4. Hacer predicciones sobre los datos de prueba
y_pred = model.predict(X_test)
```

```python
# 5. Mostrar ejemplo de predicción
print("Prediccion de Orders", y_pred[:5])
```

```python
# 6. Comparar con valores reales
print("Valores reales:", y_test[:5].values)
```

```python
# 7. Verificar importancia de las features
importances = model.feature_importances_
```

```python
# 8. Mostrar features más relevantes
feat_importances = pd.Series(importances, index=X_selected_df.columns)
print(feat_importances.sort_values(ascending=False))
```

```python
# 9. Visualizar importancia
feat_importances.sort_values().plot(kind='barh')
```

8. Cálculo de métricas básicas para validar el modelo.
 Usamos métricas de regresión: MAE, MSE, RMSE, R².

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python

# 1. Calcular MAE
mae = mean_absolute_error(y_test, y_pred)
```

```python
# 2. Calcular MSE
mse = mean_squared_error(y_test, y_pred)
```

```python
# 3. Calcular RMSE
rmse = np.sqrt(mse)
```

```python
# 4. Calcular R²
r2 = r2_score(y_test, y_pred)
```

```python
# 5. Mostrar métricas
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R2 Score: {r2:.2f}")
```

```python
# 6. Evaluar si el modelo sobreajusta
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print(f"Train R2: {train_score:.2f}, Test R2: {test_score:.2f}")
```

```python
# 7. Mostrar residuales básicos
residuals = y_test - y_pred
print("Ejemplo de residuales:", residuals[:5].values)
```

```python
# 8. Graficar predicciones vs reales (opcional)
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Valores Reales")
plt.ylabel("Predicciones")
plt.title("Predicción vs Realidad")
plt.grid(True)
plt.show()
```

9. Generar conclusiones basadas en resultados

```python
# 1. Conclusión sobre el error absoluto promedio
if mae < 10:
    print("🔎 El error absoluto medio es bajo, el modelo es preciso en general.")
else:
    print("⚠️ El modelo tiene margen de mejora en precisión.")
```

```python
# 2. Conclusión sobre R2
if r2 > 0.8:
    print("✅ Alta capacidad predictiva (R2 > 0.8).")
elif r2 > 0.5:
    print("🟡 Capacidad predictiva aceptable (R2 moderado).")
else:
    print("❌ El modelo explica poco la variabilidad (R2 bajo).")
```

```python
# 3. Comparación entre entrenamiento y prueba
gap = abs(train_score - test_score)
if gap > 0.2:
    print("⚠️ Posible sobreajuste: el modelo se ajusta demasiado a los datos de entrenamiento.")
else:
    print("✅ No hay señales fuertes de sobreajuste.")
```

```python
# 4. Features más importantes
top_feat = feat_importances.sort_values(ascending=False).head(3)
print("📌 Las 3 características más influyentes fueron:")
print(top_feat)
```
